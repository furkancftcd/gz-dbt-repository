## Json dosyalarƒ±yla √ßalƒ±≈üma

A≈üaƒüƒ±daki h√ºcreyi √ßalƒ±≈ütƒ±rarak `d` s√∂zl√ºƒü√ºn√º ba≈ülatƒ±n.

```python
d = {"firstname": "leonardo",
     "lastname": "dicaprio",
     "job": "actor"}
```

`d` klas√∂r√ºnden `actors.json` adlƒ± bir dosya olu≈üturun. S√∂zdizimi konusunda kafanƒ±z karƒ±≈ütƒ±ysa ders notlarƒ±na bakabilirsiniz.

```python
import json

with open("actors.json", "w") as f:
  json.dump(d, f)
```

Yeni `actors.json` dosyasƒ±nƒ± a√ßƒ±n ve yeni bir anahtar `nationality` ekleyin (Leonardo Dicaprio Amerikalƒ±dƒ±r). \
Sonucu aynƒ± dosya adƒ±na kaydedin. Dosyanƒ±zƒ± kontrol edin.

```python
import json  # JSON ile √ßalƒ±≈ümak i√ßin gerekli
# dosyayƒ± a√ß (okuma modu)
with open("actors.json", "r") as f:
    leonardo = json.load(f)
# milliyeti ekle
leonardo["nationality"] = "American"  # genelde b√ºy√ºk harfle yazƒ±lƒ±r
# dosyayƒ± geri kaydet (yazma modu)
with open("actors.json", "w") as f:
    json.dump(leonardo, f, indent=4)  # okunabilirlik i√ßin

```

Verileri doƒüru ≈üekilde y√ºklemek i√ßin √∂nceki kod satƒ±rƒ±nƒ±zƒ± deƒüi≈ütirin. \
≈û√ºpheleriniz varsa [belgelere](https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html) g√∂z atƒ±n.

```python
df_sales_restaurant_2=pd.read_excel("/content/sample_data/restaurants_accountability.xlsx",sheet_name="Restaurant2",header=3,usecols="C:F")
df_sales_restaurant_2
```

Her veri √ßer√ßevesine yeni bir `restaurant_id` s√ºtunu ekleyin (restaurant_1 i√ßin id = 1 ve restaurant_2 i√ßin id = 2) ve iki veri √ßer√ßevesini `df_sales` adlƒ± tek bir veri √ßer√ßevesi halinde birle≈ütirin.


ƒ∞ki veri √ßer√ßevesini birle≈ütireceƒüiniz `eksen`e dikkat edin.

```python
df_sales_restaurant_1["id"] =1
df_sales_restaurant_2["id"] =2
df_sales=pd.concat([df_sales_restaurant_1,df_sales_restaurant_2], axis=0)
df_sales
```

Birle≈ütirilmi≈ü veri √ßer√ßevesini, `All` adlƒ± bir sayfada yeni bir `restaurants_accountability_full.xlsx` dosyasƒ±na aktarƒ±n.

```python
import pandas as pd

output_filename = "restaurants_accountability_full.xlsx"

with pd.ExcelWriter(output_filename, engine='xlsxwriter', datetime_format='yyyy-mm-dd') as writer:
    df_sales.to_excel(writer, sheet_name="All", index=False)
```
